```python
# Importing packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, skew, kurtosis, yeojohnson
import numpy as np


# Read in the recipe site traffic CSV as a DataFrame
recipe = pd.read_csv("recipe_site_traffic_2212.csv", index_col='recipe')

# Display basic information about the recipe DataFrame
# This section provides an overview of the dataset
display(recipe.info())

# Display basic statistics for numerical columns
display(recipe.describe())

# Check for missing values in each column
display(recipe.isna().sum())

# Display a heatmap of missing values
plt.figure(figsize=(10, 6))
sns.heatmap(recipe.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

# Check the shape of recipe and the head
display(recipe.shape)
display(recipe.head())


# Change the data type of category to category
recipe["category"] = recipe["category"].astype("category")
display(recipe["category"].value_counts())
recipe["category"] = recipe["category"].replace({"Chicken Breast": "Chicken"})
display(recipe["category"].value_counts())


# Explore the outputs for high_traffic
display(recipe["high_traffic"].value_counts())

# Create a new option in high_traffic to be low when it is not high
recipe["high_traffic"].fillna('Low', inplace=True)
recipe['high_traffic'] = (recipe['high_traffic'] == 'High').astype(bool)
display(recipe["high_traffic"].value_counts())


#Checking and handling the servings column
# Explore the distribution of the 'serving' variable
display(recipe['servings'].value_counts())

# Remove 'as a snack' from the 'servings' column
recipe['servings'] = pd.to_numeric(recipe['servings'].str.replace(' as a snack', ''))

# Convert 'servings' column to numeric
recipe['servings'] = pd.to_numeric(recipe['servings'])

# Display the updated distribution of 'servings'
display(recipe['servings'].value_counts())


# Drop duplicates
recipe.drop_duplicates(inplace= True)
display(recipe.shape)

# Check for missing values in each column
display(recipe.isna().sum())

# Drop rows with missing values below a 5% threshold
threshold = len(recipe) * 0.05
cols_to_drop = recipe.columns[recipe.isna().sum() <= threshold]
recipe.dropna(subset=cols_to_drop, inplace=True)

# Check for missing values in each column after dropping rows
display(recipe.isna().sum())


# List of the numeric variable calories, carbohydrate, sugar nd protein
numeric_columns = ['calories', 'carbohydrate', 'sugar', 'protein']

# check if the numeric variables have negative values
for col in numeric_columns:
  if recipe[col].min() < 0:
    print(f'Column {col} contains negative values')
  else:
    print(f'Column {col} contains no negative values')

# Plot histograms for each numeric column
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)
    sns.histplot(recipe[col], kde=True, bins=30, color='blue')
    plt.title(f'Histogram of {col}')

plt.tight_layout()
plt.show()


# Calculate skewness and kurtosis for each numeric column
for col in numeric_columns:
    skewness = skew(recipe[col].dropna())
    print(f"Skewness of '{col}': {skewness}")
for col in numeric_columns:
    kurt = kurtosis(recipe[col].dropna())
    print(f"Kurtosis of '{col}': {kurt}")
    

# Log-transform numeric columns
for col in numeric_columns:
    recipe[col] = np.log1p(recipe[col])
    

 # Calculate skewness and kurtosis for each numeric column
for col in numeric_columns:
    skewness = skew(recipe[col].dropna())
    print(f"Skewness of '{col}': {skewness}")
for col in numeric_columns:
    kurt = kurtosis(recipe[col].dropna())
    print(f"Kurtosis of '{col}': {kurt}")
    
# Check if transformed data and without outliers is normally distributed
def check_normality(data):
    # Plot the histogram
    plt.hist(data, bins=30, density=True, alpha=0.6, color='blue')
    # Fit a normal distribution to the data
    mu, std = norm.fit(data)
    # Plot the normal distribution as a red line
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'r', linewidth=2)
    plt.text(0.5, -0.25, "Fit Results: Mean = %.2f, Std = %.2f" % (mu, std), ha='center', va='center', transform=plt.gca().transAxes, fontsize=10)

for i, col in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)
    check_normality(recipe[col])
    plt.title(f'Normality of {col}')
plt.tight_layout()
plt.show()

# Function to drop outliers based on IQR for specified columns
def drop_outliers(columns, dataframe):
    for col in columns:
        Q1 = dataframe[col].quantile(0.25)
        Q3 = dataframe[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = dataframe[(dataframe[col] < lower_bound) | (dataframe[col] > upper_bound)]
        dataframe = dataframe.drop(outliers.index, axis=0)

    return dataframe

# Create boxplot before handling outliers
plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 1)
sns.boxplot(data=recipe[numeric_columns])
plt.title('Boxplot Before Handling Outliers')

# Drop outliers in the transformed data
recipe_no_outliers = drop_outliers(numeric_columns, recipe.copy())

# Create boxplot after handling outliers
plt.subplot(1, 2, 2)
sns.boxplot(data=recipe_no_outliers[numeric_columns], showfliers=False)
plt.title('Boxplot After Handling Outliers')


plt.tight_layout()
plt.show()

# Set up the matplotlib figure and check normailty after outliers
plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)
    check_normality(recipe_no_outliers[col])
    plt.title(f'Normality of {col}')
plt.tight_layout()
plt.show()



# Get dummies for 'category' and keep the original column
recipe_dummies = pd.get_dummies(recipe_no_outliers, columns=['category'], prefix='category')

# Replace spaces in dummy column names with underscores
recipe_dummies.columns = recipe_dummies.columns.str.replace(' ', '_')

# Concatenate the original DataFrame and the dummy columns
recipe_encoded = pd.concat([recipe_no_outliers, recipe_dummies], axis=1)
recipe_encoded = recipe_encoded.loc[:, ~recipe_encoded.columns.duplicated()]

# Convert 'category' column to categorical type
recipe_encoded['category'] = recipe_encoded['category'].astype('category')


# Check for missing values
display(recipe_encoded.isna().sum())
print(recipe_encoded.info())
```

The goal is to build a model that accurately predicts recipe popularity to enhance the user experience in a recipe recommendation platform. The problem at hand is a binary classification task, aiming to categorize recipes into high-traffic and low-traffic. 

Firstly, different supervised machine learning models have been selected to compare their results.
- Logistic Regression: A simple and interpretable model often used for binary classification tasks, suitable for establishing a baseline.
- Random Forest: An ensemble method known for its ability to handle complex relationships in the data, offering robust performance.
- Support Vector Machine (SVM): Effective for both linear and non-linear classification tasks, providing flexibility in capturing diverse patterns.
- Gradient Boosting: A boosting algorithm that builds decision trees sequentially, enhancing predictive performance over iterations.
- Neural Network: A deep learning model capable of capturing intricate patterns in the data, suitable for complex relationships.

Data processing was carried out. 
- Data extraction was carried out: Extracted features from the dataset, excluding irrelevant columns (recipe, high_traffic, category).
- Oversampling: Applied oversampling to the minority class using RandomOverSampler to address class imbalance.
- Train-Test Split: Split the resampled data into training and testing sets (80% training, 20% testing) for model evaluation.
- Preprocessing Steps: Defined a preprocessor using StandardScaler for numerical feature standardization.

Model Development was then completed by creating pipelines for each model, incorporating the defined preprocessor and the respective classification model. The model then had Logistic Regression, Random Forest, SVM, Gradient Boosting, and Neural Network models with default hyperparameters for the initial model development.
Lastly a pipeline creation was made to streamline the preprocessing and modeling steps for each selected model.

```python
# Import relevant packages
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

# Set a random seed
random_seed = 42

# Split the data into features (X) and target variable (y)
X = recipe_encoded.drop(['high_traffic', 'category'], axis=1)
y = recipe_encoded['high_traffic']

# Apply oversampling to the minority class
oversampler = RandomOverSampler(sampling_strategy='minority', random_state=random_seed)
X_resampled, y_resampled = oversampler.fit_resample(X, y)

# Split the resampled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=random_seed)

# Define preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), X.columns)
    ])

# Define models and their hyperparameter search spaces
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'Neural Network': MLPClassifier()
}

# Create pipelines
pipelines = {}
for model_name, model in models.items():
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    pipelines[model_name] = pipeline
```

The evaluation includes printing the classification report, accuracy, and cross-validation scores for each model. The metrics used for evaluation are precision, recall, F1-score, and accuracy.
The code used different parameter:
- Hyperparameter Tuning:
    - It uses GridSearchCV to perform hyperparameter tuning for each model based on the specified parameter grids (param_grids).
    - The best estimator for each model is stored in the best_estimators dictionary.
- Model Evaluation:
    - For each best-tuned model, it prints the performance metrics on the test set.
    - The classification report includes precision, recall, and F1-score for both classes (True and False).
    - Accuracy is also printed.
    - Cross-validation scores are calculated using cross_val_score with a 5-fold StratifiedKFold.
- Metrics:
    - Precision: The ability of the classifier not to label as positive a sample that is negative.
    - Recall: The ability of the classifier to find all the positive samples.
    - F1-score: The weighted average of precision and recall.
    - Accuracy: The ratio of correctly predicted instances to the total instances.
- Cross-Validation:
    - Provides an estimate of how well the model will perform on an independent dataset.
- Confidence Interval:
    - Provides Confidence Interval to give a more comprehensive view of the model's stability and generalization performance. 


Note. The code below takes a few minutes to process.

```python
# Import relevant packages
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.model_selection import GridSearchCV
from imblearn.pipeline import make_pipeline

# Hyperparameter search spaces
param_grids = {
    'Logistic Regression': {'model__C': [0.1, 1, 10]},
    'Random Forest': {'model__n_estimators': [50, 100, 150], 'model__max_depth': [3, 5, 7]},
    'SVM': {'model__C': [0.1, 1, 10], 'model__kernel': ['linear', 'rbf']},
    'Gradient Boosting': {'model__n_estimators': [50, 100, 150], 'model__learning_rate': [0.01, 0.1, 0.2], 'model__max_depth': [3, 5, 7]},
    'Neural Network': {'model__hidden_layer_sizes': [(50,), (100,), (150,)], 'model__max_iter': [100, 200, 300]}
}

# Perform hyperparameter tuning
best_estimators = {}
for model_name, pipeline in pipelines.items():
    param_grid = param_grids[model_name]
    grid_search = GridSearchCV(pipeline, param_grid, cv=StratifiedKFold(n_splits=3), scoring='accuracy')
    grid_search.fit(X_train, y_train)
    best_estimators[model_name] = grid_search.best_estimator_

# Evaluate models
for model_name, pipeline in best_estimators.items():
    print(f"{model_name} Performance:")
    y_pred = pipeline.predict(X_test)
    print(classification_report(y_test, y_pred))
    print("Accuracy:", accuracy_score(y_test, y_pred))

    # Cross-Validation
    cv_scores = cross_val_score(pipeline, X_resampled, y_resampled, cv=StratifiedKFold(n_splits=5), scoring='accuracy')

    # Calculate confidence interval
    mean_accuracy = np.mean(cv_scores)
    std_accuracy = np.std(cv_scores)
    confidence_interval = 1.96 * (std_accuracy / np.sqrt(len(cv_scores)))

    print(f"\nCross-Validation Scores for {model_name}:")
    print(cv_scores)
    print("Mean Accuracy:", mean_accuracy)
    print("Confidence Interval:", confidence_interval)
    print("\n")
```
